```python
x_train =df[['gender', 'age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi','smoking_status']] 
```

```python
x_train = x_train.reset_index()
```

```python
#normalising age,avg_glucose_level,bmi 
x_train["age"] = (x_train["age"]-x_train["age"].min())/(x_train["age"].max()-x_train["age"].min())
x_train["avg_glucose_level"] = (x_train["avg_glucose_level"]-x_train["avg_glucose_level"].min())/(x_train["avg_glucose_level"].max()-x_train["avg_glucose_level"].min())
x_train["bmi"] = (x_train["bmi"]-x_train["bmi"].min())/(x_train["bmi"].max()-x_train["bmi"].min())
```

```python
x_train.drop("index",axis=1,inplace=True)
```

```python
y_train = df["stroke"]
```

```python
# we know that linear regression= sigmoid = = 1/1+e^-z
#where z = w.x + z 
def sigmoid(z):
    g= 1/(1+np.exp(-z))
    
    return g
```

```python
sigmoid(0)
```

```python
#calculating cost for the model 
def comp_cost(x,y,w,b):
    
    m,n = x.shape
    print("m",m)
    print("n",n)
    z_wb =0
    loss_sum=0
    epsilon = 1e-15 
    
    for i in range(m):
        for j in range(n):
            z_wb_j =  w[j]*x.iloc[i][j] 
            z_wb+=z_wb_j
        z_wb+=b
        #print(z_wb)
        func = sigmoid(z_wb)
        #print(func)
        loss = (-y.iloc[i] * np.log(func + epsilon)) - (1 - y.iloc[i]) * np.log(1 - func + epsilon)
        #print("loss",loss)
        loss_sum+= loss
        #print("loss_sum",loss_sum)
        #print("i",i)
    cost = loss_sum/m
    print("cost",cost)
    return cost
    
```

```python
n = 7
# Compute and display cost with w and b initialized to zeros
initial_w = np.array([1,2,3,4,5,6,7])
initial_b = 0
cost = comp_cost(x_train, y_train, initial_w, initial_b)
print('Cost at initial w and b (zeros): {:.3f}'.format(cost))
```
